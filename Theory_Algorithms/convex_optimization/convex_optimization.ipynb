```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_1=np.array([1,0],dtype=np.float32)
x_2=np.array([0,1],dtype=np.float32)
X=[]
Y=[]
for i in np.arange(-3,3,0.01):
    x=x_1*i+x_2*(1-i)
    X.append(x[0])
    Y.append(x[1])
plt.scatter(X,Y)  
```
